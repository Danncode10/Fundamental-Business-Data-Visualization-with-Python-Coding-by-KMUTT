### Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Basic line chart: service vs salary
plt.figure(figsize=(8,5)) # width: 8, height: 5
plt.plot(phd)
plt.plot(service)

# Add titles and labels
plt.title("Title of the graph")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.grid(True)

plt.show()
```

```python
categories = ['Dog', 'Cat', 'Rabbit']
values = [5, 3, 4]

plt.bar(categories, values, color='green')
plt.title("Number of Animals")
plt.xlabel("Animal")
plt.ylabel("Count")
plt.show()
```

```python
# Example data
animals = ['Dog', 'Cat', 'Rabbit', 'Parrot']
counts = [5, 3, 2, 1]

plt.figure(figsize=(6,6))
plt.pie(counts, labels=animals)
plt.title("Animal Distribution")
plt.show()
```

```python
# Example data
hours_studied = [1, 2, 3, 4, 5]
scores = [50, 55, 65, 70, 90]

plt.figure(figsize=(6,4))
plt.scatter(hours_studied, scores, color='red', s=100, marker='o')
plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

```python
# Sample data
import pandas as pd
df = pd.DataFrame({
    "Animal": ["Dog","Cat","Rabbit","Dog","Cat","Rabbit"],
    "Age": [5, 3, 1, 6, 4, 2]
})

plt.figure(figsize=(6,4))
sns.boxplot(x="Animal", y="Age", data=df)
plt.title("Age Distribution by Animal")
plt.show()
```

```python
# Sample data
df = pd.DataFrame({
    "Hours_Studied": [1, 2, 3, 4, 5, 6, 7],
    "Exam_Score": [50, 55, 65, 70, 80, 85, 90]
})

# Create joint plot
sns.jointplot(x="Hours_Studied", y="Exam_Score", data=df, kind="scatter")
plt.show()
```
