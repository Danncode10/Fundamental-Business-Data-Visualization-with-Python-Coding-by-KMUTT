```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the dataframe
df = pd.DataFrame({
    "Name": ["Ahmed", "Omar", "Ali", "Ziad", "Salwa", "Lila"],
    "Mobile_Sales": [2540, 1370, 1320, 2000, 2100, 2150],
    "TV_Sales": [2200, 1900, 2150, 1850, 1770, 2000]
})

name = df["Name"]
mobile_sales = df["Mobile_Sales"]
tv_sales = df["TV_Sales"]

df
```

### a. Create a bar plot of the sales volume.

```python
# Mobile sales
plt.figure(figsize=(8,5))
plt.bar(name, mobile_sales, color='blue', label='Mobile Sales')
plt.title("Mobile Sales")
plt.xlabel("Name")
plt.ylabel("Sales")
plt.legend()
plt.show()


# Tv sales
plt.figure(figsize=(8,5))
plt.bar(name, tv_sales, color='orange', label='TV Sales') 
plt.title("TV Sales")
plt.xlabel("Name")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

### b. Pie chart of total item sales

```python
total_sales = [mobile_sales.sum(), tv_sales.sum()]
text_labels = ['Mobile', 'TV']

plt.figure(figsize=(6,6))
plt.pie(total_sales, labels=text_labels, autopct='%1.1f%%', colors=['blue','orange'])
plt.title("Total Sales Distribution")
plt.show()
```

### c. Box plot of item sales

```python
plt.figure(figsize=(6,4))
plt.boxplot([mobile_sales, tv_sales], tick_labels=['Mobile', 'TV'])
plt.title("Box Plot of Sales")
plt.ylabel("Sales")
plt.show()
```

### d. Area plot of item sales

```python
df_area = df.set_index('Name')
df_area.plot(kind='area')
plt.title("Area Plot of Sales")
plt.ylabel("Sales")
plt.show()
```

### e. Stacked bar plot of item sales

```python
df_area.plot(kind='bar', stacked=True, figsize=(8,5), color=['blue','orange'])
plt.title("Stacked Bar Plot of Sales")
plt.ylabel("Sales")
plt.show()
```
